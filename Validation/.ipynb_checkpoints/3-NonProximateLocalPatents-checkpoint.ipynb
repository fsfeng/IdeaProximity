```python
rs = fastparquet.ParquetFile("DataStore/2018-11/primclass_sim_claims_1120.parq").to_pandas()
```

```python
rs_sl = rs.loc[rs["inv_msa_match"] == True]
rs_dl = rs.loc[rs["inv_msa_match"] == False]
```

```python
%matplotlib inline
sns.set(font_scale=2)
sns.set_style("ticks")
plt.rcParams["font.family"] = "Liberation Sans"
plt.rcParams.update({'font.size': 14})
plt.style.use('seaborn-muted')
```

```python
# Abstract sim
col = "sim_docvecs"
fig, ax = plt.subplots()
x = rs_sl[col]*100
y = rs_dl[col]*100
for a in [x,y]:
    sns.distplot(a, bins=range(-21,81,10), ax=ax, norm_hist=True, kde=False)

print(np.round(rs_sl[col].describe(),2),np.round(rs_dl[col].describe(),2))
```

```python
# Abstract sim
col = "sim_claims_docvecs"
fig, ax = plt.subplots()
x = rs_sl[col]*100
y = rs_dl[col]*100
for a in [x,y]:
    sns.distplot(a, bins=range(-21,81,10), ax=ax, norm_hist=True, kde=False)

print(np.round(rs_sl[col].describe(),2),np.round(rs_dl[col].describe(),2))
```

```python
# Abstract sim
col = 'num_shared_claims_stemmed'
fig, ax = plt.subplots()
x = rs_sl[col]
y = rs_dl[col]
for a in [x,y]:
    sns.distplot(a, bins=range(0,20,1), ax=ax, norm_hist=True, kde=False)

print(np.round(rs_sl[col].describe(),2),np.round(rs_dl[col].describe(),2))
```
